```python
data_filtered = data[data['mp_25'].notna()].drop('p_atm', axis=1).dropna()
data_filtered.head()
```

```python
def f_season(row, componente):
    dict_map = {'otoño': (1, 0), 
            'invierno': (0, 1), 
            'primavera': (-1, 0), 
            'verano': (0, -1)}
    
    season_cos, season_sin = dict_map[row['Estacion']]
    if componente == 'sin':
        return season_sin
    if componente == 'cos':
        return season_cos
```

```python
data_filtered['season_cos'] = data_filtered.apply(f_season, axis=1, componente='cos')
data_filtered['season_sin'] = data_filtered.apply(f_season, axis=1, componente='sin')

# nuevo df que va a correr las horas
mp_25 = data_filtered[['datetime', 'mp_25']]
mp_25.set_index('datetime', drop=True, inplace=True)
mp_25.reset_index(inplace=True)

# corremos las horas
dt = timedelta(hours=1)
mp_25.loc[:,'datetime'] -= dt

# añadiendo mp_25 corrido al dataframe filtered
data_filtered.drop('mp_25', axis=1, inplace=True)
data_filtered = pd.merge(data_filtered, mp_25, on='datetime')

# botando lo que no necesitamos
data_filtered.drop(['Estacion', 'datetime'], axis=1, inplace=True)
data_filtered.head()
```

```python
sb.pairplot((data_filtered - data_filtered.mean())/data_filtered.std(), corner=True)
```

```python
data_std = (data_filtered - data_filtered.mean()) / data_filtered.std()
```

```python
X = data_std.drop('mp_25', axis=1).values
y = data_filtered['mp_25'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Parte con Torch

```python
class MP25Dataset(Dataset):
    """Dataset para el proyecto"""
    def __init__(self, X, y):
        self.X = torch.from_numpy(X)
        self.y = torch.from_numpy(y)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        sample = self.X[idx], self.y[idx]
        return sample
    def __repr__(self):
        return 'MP25Dataset'
```

```python
train_dataset = MP25Dataset(X_train, y_train)
test_dataset = MP25Dataset(X_test, y_test)
```

```python
class Net(nn.Module):
    def __init__(self):
        
        # FF
        self.fc1 = nn.Linear(12, 10)
        self.fc2 = nn.Linear(10, 50)
        self.fc3 = nn.Linear(50, 100)
        self.fc4 = nn.Linear(100, 100)
        self.fc5 = nn.Linear(100, 50)
        self.fc6 = nn.Linear(50, 1)
        
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = F.relu(self.fc5(x))        
        x = F.relu(self.fc6(x))

        return x
```

```python
device = ('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
net = Net()
net.to(device)
```

```python
loss_func = torch.nn.MSELoss().to(device)
opt = optim.Adam(net.parameters(), weight_decay=1)
```

```python
for epoch in range(3):  # loop over the dataset multiple times
    running_loss = 0.0
    for i in range(len(train_dataset)):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = train_dataset[i]

        # zero the parameter gradients
        opt.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs.float().to(device))
        loss = loss_func(outputs.double(), labels.double()*10).to(device)
        loss.backward()
        opt.step()

        # print statistics
        running_loss += loss.item()
        if i % 2000 == 1999:    # print every 2000 mini-batches
            print(outputs.double(), labels.double()*10)
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0
```
